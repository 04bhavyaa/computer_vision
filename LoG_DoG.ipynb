```python
# compute laplacian of gaussian (LoG), comment on results after changing params
```

```python
import cv2
import matplotlib.pyplot as plt

image_path = 'image.jpg'
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Display the grayscale image
    plt.imshow(gray_image, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off') # Hide axes
    plt.show()
```

```python
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Compute Laplacian
log_image = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Display the LoG image
plt.imshow(log_image, cmap='gray')
plt.title("Laplacian of Gaussian (LoG) with Default Parameters")
plt.axis('off')  # Turn off axes
plt.show()
```

```python
# Apply Gaussian blur with kernel size (9, 9) and sigma 2
blurred_image_param1 = cv2.GaussianBlur(gray_image, (9,9), 2)

# Compute Laplacian
log_image_param1 = cv2.Laplacian(blurred_image_param1, cv2.CV_64F)

# Display the LoG image with parameters
plt.imshow(log_image_param1, cmap='gray')
plt.title("Laplacian of Gaussian (LoG) with Kernel (9, 9), Sigma 2")
plt.axis('off')
plt.show()
```

### Difference of Gaussian

```python
# Apply Gaussian blur with sigma 0
blurred_image_1 = cv2.GaussianBlur(gray_image, (5, 5), 0.6)

# Apply Gaussian blur with sigma 1
blurred_image_2 = cv2.GaussianBlur(gray_image, (5, 5), 1.2)

# Compute the difference of Gaussians
dog_image_initial = cv2.subtract(blurred_image_1, blurred_image_2)

# Display the DoG image
plt.imshow(dog_image_initial, cmap='gray')
plt.title("Difference of Gaussians (DoG)")
plt.axis('off')  # Turn off axes
plt.show()
```

```python
# Apply Gaussian blur with sigma 2
blurred_image_sigma2 = cv2.GaussianBlur(gray_image, (5, 5), 2)

# Apply Gaussian blur with sigma 3
blurred_image_sigma3 = cv2.GaussianBlur(gray_image, (5, 5), 3)

# Compute the difference of Gaussians (sigma 2 and sigma 3)
dog_image_sigma2_sigma3 = cv2.subtract(blurred_image_sigma2, blurred_image_sigma3)

# Display the DoG image with sigma 2 and sigma 3
plt.figure(figsize=(10, 5)) # Adjust figure size for two plots
plt.subplot(1, 2, 1) # First subplot
plt.imshow(dog_image_sigma2_sigma3, cmap='gray')
plt.title("DoG (Sigma 2 and Sigma 3)")
plt.axis('off')

# Apply Gaussian blur with sigma 1
blurred_image_sigma1 = cv2.GaussianBlur(gray_image, (5, 5), 1)

# Apply Gaussian blur with sigma 5
blurred_image_sigma5 = cv2.GaussianBlur(gray_image, (5, 5), 5)

# Compute the difference of Gaussians (sigma 1 and sigma 5)
dog_image_sigma1_sigma5 = cv2.subtract(blurred_image_sigma1, blurred_image_sigma5)

# Display the DoG image with sigma 1 and sigma 5
plt.subplot(1, 2, 2) # Second subplot
plt.imshow(dog_image_sigma1_sigma5, cmap='gray')
plt.title("DoG (Sigma 1 and Sigma 5)")
plt.axis('off')

plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```
